```python
# create a simple text file

import os
os.makedirs('../data/text_files',exist_ok=True)
```

```python
#
sample_texts = {
    '../data/text_files/python_intro.txt': """
Python is a high-level, interpreted programming language known for its simplicity and readability.
It is widely used in web development, data science, machine learning, automation, and scripting.
Python uses indentation to define code blocks, making it visually clean and easy to understand.

Basic features of Python include dynamic typing, automatic memory management, and a large standard library.
It supports multiple programming paradigms such as procedural, object-oriented, and functional programming.

Example:
print("Hello, World!")
    """,

    '../data/text_files/machine_learning_basics.txt': """
Machine learning is a subset of artificial intelligence that enables systems to learn from data. 
It involves training models on datasets so they can make predictions or decisions without being explicitly programmed. 
Common types include supervised learning, unsupervised learning, and reinforcement learning.
    """
}

for filepath,content in sample_texts.items():
    with open(filepath,'w',encoding= 'utf-8') as f:
        f.write(content)


# here file path is =  ../data/text_files/python_intro.txt 

print('sample text files created')
```

```python
### Text loader

# from langchain.document_loaders import TextLoader

from langchain_community.document_loaders import TextLoader

```

```python
loader = TextLoader("../data/text_files/python_intro.txt", encoding = 'utf-8')

loader
```

```python
document = loader.load()

print(document)
```

```python
# directory loader to load all the text files in the directory

from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "../data/text_files",
    glob = "**/*.txt", # glob is to find pattern to match files
    loader_cls= TextLoader, # which type of file loader class to use
    show_progress= False
)

documents = dir_loader.load()
documents
```

# for pdf

```python
from langchain_community.document_loaders import PyPDFLoader,PyMuPDFLoader

dir_pdf_loader = DirectoryLoader(
    "../data/pdf",
    glob = "**/*.pdf",
    loader_cls= PyMuPDFLoader,
    # no encoding requiired for pdf  by default
    show_progress= False
)

doc_pdf = dir_pdf_loader.load()

doc_pdf

# PyMupdf is better than pypdf, i
# it give almost all type of metadata along with the page content in document structure
```
